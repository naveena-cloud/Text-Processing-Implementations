```python
# Import ngrams from nltk
from nltk import ngrams

# ---------- Input ----------
sentence = "Knowledge grows with practice"
n = 2 # Change value of n as needed

# ---------- Word N-grams ----------
words = sentence.split()
word_ngrams = ngrams(words, n)

print("Word N-grams:")
for gram in word_ngrams:
    print(gram)

# ---------- Character N-grams ----------
char_ngrams = ngrams(sentence, n)

print("\nCharacter N-grams:")
for gram in char_ngrams:
    print("".join(gram))
```
