```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize

# Run once if required
# nltk.download('punkt')
# nltk.download('stopwords')
# nltk.download('wordnet')

text = "The workers are processing and analyzing the data"

# Tokenization
words = word_tokenize(text)

# Stopword Removal
stop_words = set(stopwords.words('english'))
filtered_words = [w for w in words if w.lower() not in stop_words]

# Lemmatization
lemmatizer = WordNetLemmatizer()
lemmatized_words = [
    lemmatizer.lemmatize(w, 'v') if w.endswith('ing')
    else lemmatizer.lemmatize(w)
    for w in filtered_words
]

print("Original Text:", text)
print("After Stopword Removal:", filtered_words)
print("After Lemmatization:", lemmatized_words)
```
