```python
from collections import Counter

text = "Deep learning algorithms learn complex patterns and learn useful representations"
w = text.lower().split()

# Bigram (context) and trigram counts
bi = Counter(zip(w, w[1:]))
tri = Counter(zip(w, w[1:], w[2:]))

V = len(set(w))  # vocabulary size

# Laplace smoothed trigram probability
P = lambda w1, w2, w3: (tri[(w1, w2, w3)] + 1) / (bi[(w1, w2)] + V)

print(P("deep", "learning", "algorithms"))
print(P("patterns", "and", "future"))
```
